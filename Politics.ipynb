```python
# importing
import pandas as pd
import numpy as np

# read data
pol_data = pd.read_csv("politics.csv")
data = pd.read_csv("stage3.csv")
pop_data = pd.read_csv("gun_ownership.csv")
```

```python
# more
from bokeh.io import output_notebook, show
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.palettes import viridis
```

```python
# counting all the percentages together
# not usefull at all, but done

counter = 0
for i in range(1, len(pol_data["Democrat"])):    
    counter += pol_data["Democrat"][i]
print(counter, "the percentage of democratic people")

counter = 0
for i in range(1, len(pol_data["Republican"])):
    counter += pol_data["Republican"][i]
print(counter, "the percentage of replublican people")

counter = 0
for i in range(1, len(pol_data["Other"])):
    counter += pol_data["Other"][i]
print(counter, "the percentage of other people")
```

```python
# per staat het aantal democraten
P_S = {}

for i in range(len(pop_data["population"])):
    if pop_data["state"][i] not in P_S:
        P_S[pop_data["state"][i]] = pop_data["population"][i]

# percentage voorkeur per staat
D_S = {}

for i in range(len(pol_data["Democrat"])):
    if pol_data["State"][i] not in D_S:
        D_S[pol_data["State"][i]] = pol_data["Democrat"][i]

VForVis = []
USState = []    
    
for state in P_S:
    USState.append(state)
    VForVis.append(int(P_S[state]/100*D_S[state]))
    
# creating the histogram
output_file("Democratic_people.html")

p = figure(x_range=USState, plot_height=250, title="Democratic people per state", 
          toolbar_location=None, tools="")

p.vbar(x=USState, top=VForVis, width=0.5)
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
# per staat het aantal republikeinen
P_S = {}

for i in range(len(pop_data["population"])):
    if pop_data["state"][i] not in P_S:
        P_S[pop_data["state"][i]] = pop_data["population"][i]

# percentage voorkeur per staat
R_S = {}

for i in range(len(pol_data["Republican"])):
    if pol_data["State"][i] not in R_S:
        R_S[pol_data["State"][i]] = pol_data["Republican"][i]

VForVis = []
USState = []    
    
for state in P_S:
    USState.append(state)
    VForVis.append(int(P_S[state]/100*R_S[state]))
    
# creating the histogram
output_file("Republican_people.html")

p = figure(x_range=USState, plot_height=250, plot_width=2500, title="Republican people per state", 
          toolbar_location=None, tools="")

p.vbar(x=USState, top=VForVis, width=0.5)
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```
